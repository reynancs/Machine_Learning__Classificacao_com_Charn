## <font color=blue>2.1 Transformando as Variáveis Categóricas em Numéricas </font>

### <font color=blue> Categorias Multivariáveis </font>

```python
# Habilitar para vê o número total de colunas da tabela
pd.set_option('display.max_columns',39)
```

```python
categorias_multi = ['SegurancaOnline', 'VariasLinhasTelefonicas','ServicoDeInternet',
                    'BackupOnline','SeguroNoDispositivo','SuporteTecnico','TVaCabo',
                    'StreamingDeFilmes','TipoDeContrato', 'FormaDePagamento']
```

#### Ex01) get_dummies( )
O método get_dummies() é simples de utilizar e faz a transformação de forma direta das variáveis categóricas. Não se aplica para dados novos

```python
dados_transformados = pd.get_dummies(dados, columns= categorias_multi)
dados_transformados.head()
```

#### Ex02) OneHotEncoder ( )
Já o método OneHotEncoder() funciona como outros modelos disponíveis na biblioteca Sklearn. Nele, é necessário instanciar um objeto e depois ajustar aos dados com um método fit(). Dessa forma, esse objeto armazena os passos necessários para realizar a transformação dos dados.

```python
'''from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

one_hot_enc = make_column_transformer((OneHotEncoder(handle_unknown = 'ignore'),categorias_multi),remainder='passthrough')

dados_transformados = one_hot_enc.fit_transform(dados)
dados_transformados = pd.DataFrame(dados_transformados, columns=one_hot_enc.get_feature_names_out())
dados_transformados.head()'''
```

### <font color=blue> Categorias Binárias </font>

```python
traducao_dic = {
    'Sim':1,
    'Nao':0
}

dados_transformados= dados_transformados.replace(traducao_dic)
dados_transformados.head()
```

```python
# Para podermos aplicar o SMOTE, devemos separar  os dados em variáveis características e resposta  

x = dados_transformados.drop('Churn', axis= 1)
y = dados_transformados[['Churn']]
```

## <font color=blue>2.2. Balanceamento dos Dados</font>

```python
import seaborn as sns
%matplotlib inline

ax = sns.countplot(x='Churn', data=dados_transformados)
```

Analisando o gráfico, podemos perceber que nossa variável está desbalanceada e precisamos tratar isso. Para realizar o balanceamento dos dados, utilizaremos uma técnica chamada de ***Oversampling***, que consiste em realizar a criação de novas observações da classe quando há menos amostras, tendo como objetivo igualar a proporção entre as categorias.

### <font color=blue> Aplicando Oversampling - SMOTE( )</font>

Uma das técnicas de oversampling muito utilizada é a **SMOTE**. Sua ideia consiste em criar observações intermediárias entre os dados que estão próximos.

```python
#!pip install -U imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE

smt = SMOTE(random_state=123)  # Instancia um objeto da classe SMOTE
x, y = smt.fit_resample(x, y)  # Realiza a reamostragem do conjunto de dados
```

```python
dados_transformados = pd.concat([x, y], axis=1)  # Concatena a variável target (y) com as features (X) 

# Verifica se o balanceamento e a concatenação estão corretos.
dados_transformados.head()
```

```python
ax = sns.countplot(x='Churn', data=dados_transformados)  # plotando a variável target balanceada.
```

##  <font color=blue> 2.3. Normalização dos Dados</font>

Antes de Aplicar o método KNN é preciso normalizar os dados, colocando todos na mesma base

```python
#biblioteca para Normalizar os Dados (dataset)
from sklearn.preprocessing import StandardScaler

norm = StandardScaler()
x_norm= norm.fit_transform(x)
x_norm
```

***
# <font color=blue>3. Implementando Modelo</font>
***

```python
#biblioteca para divisão dos dados
from sklearn.model_selection import train_test_split

x_train,x_test, y_train, y_test = train_test_split(x_norm, y, test_size=0.3, random_state=123)
```

## <font color=blue>3.1.  Modelo KNN</font>

```python
#biblioteca para criarmos o modelo de machine learning
from sklearn.neighbors import KNeighborsClassifier

#instanciar o modelo (criamos o modelo) - por padrão são 5 vizinhos  
knn = KNeighborsClassifier(metric='euclidean')
knn.fit(x_train, y_train) #treinando o modelo com os dados de treino
predito_knn= knn.predict(x_test) #testando o modelo com os dados de teste
```

## <font color=blue>3.2. Modelo BernoulliNB</font> 

```python
# Escolha utilizar a mediana, porque é o valor central dos nossos dados ordenados
np.median(x_train)
```

```python
from sklearn.naive_bayes import BernoulliNB

bnb= BernoulliNB(binarize= -0.44) # significa que: todo valor acima de -0.44 se tornará = 1 e abaixo = 0 (Binário)
bnb.fit(x_train, y_train)
predito_bnb= bnb.predict(x_test)
```

## <font color=blue>3.3. Modelo DecisionTreeClassifier</font>  

```python
#biblioteca para criarmos o modelo de machine learning
from sklearn.tree import DecisionTreeClassifier

#instanciando o modelo
dtc = DecisionTreeClassifier(criterion='entropy', random_state=42)

#treinar o modelo
dtc.fit(x_train, y_train)

predito_ArvoreDecisao = dtc.predict(x_test)
```

***
# <font color=blue>4. Validação dos Modelos</font>
***

## <font color=blue> 4.1. Matriz de Confusão</font>  

image.png

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, predito_knn))

print(confusion_matrix(y_test, predito_bnb))

print(confusion_matrix(y_test, predito_ArvoreDecisao))
```

## <font color=blue> 4.2. Acurácia </font>

No nosso modelo a acurácia mede ***o quanto o nosso modelo acertou do total, do total de resultados o quanto que ela acertou.***

A partir do calculo da matriz de confusão conseguimos inferir outras métricas, como por exemplo a acurária.

$ACC$ = ${TP + TN \over TP + FP + TN + FN}$

```python
from sklearn.metrics import accuracy_score

#modelo KNN
print(accuracy_score(y_test, predito_knn))

#modelo Bernoulli de naive bayes
print(accuracy_score(y_test, predito_bnb))

#modelo árvore de decisão
print(accuracy_score(y_test, predito_ArvoreDecisao))

```

## <font color=blue> 4.3. Precisão</font>

Outra métrica importante é a precisão, que calcula ***quantos foram classificados corretamento como positivos ($TP$).***

$PS$ = ${TP \over TP + FP}$


```python
from sklearn.metrics import precision_score

#modelo KNN
print(precision_score(y_test, predito_knn))

#modelo Bernoulli de naive bayes
print(precision_score(y_test, predito_bnb))

#modelo árvore de decisão
print(precision_score(y_test, predito_ArvoreDecisao))
```

## <font color=blue> Recall </font>

 Mede o ***quão bom o nosso modelo está em classificar os resultados realmente positivos.***

Outra métrica importante é a Recall ou revocação ou ainda sensibilidade, que calcula o quão bom o modelo está para classificar corretamente um resultado positivo ($TP$).

$RC$ = ${TP \over TP + FN}$

```python
from sklearn.metrics import recall_score

#modelo KNN
print(recall_score(y_test, predito_knn))

#modelo Bernoulli de naive bayes
print(recall_score(y_test, predito_bnb))

#modelo árvore de decisão
print(recall_score(y_test, predito_ArvoreDecisao))
```

Agora, ao final deste case conseguimos responder o problema de negócio proposto: Qual modelo classifica melhor nossos clientes?
Neste estudo foi decidido considerar a métrica de precisão que mede quantos valores positivos foram previstos de forma correta com um todo no nosso modelo. Lembrando que os verdadeiros positivos é o foco, aquele onde Churn =1. Sendo assim, é este Churn que precisamos dar atenção, é o Churn que temos que reduzir, são clientes que estão deixando a empresa, reduzindo seus ganhos.
